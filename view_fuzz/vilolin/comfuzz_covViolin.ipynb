```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 加载示例数据集
# df = pd.read_csv('data/comfuzz_test.csv')
df = pd.read_csv('data/comfuzz_all.csv')
df['Cover'] = df['Cover'].str.replace('%','').astype(float)
df['Cover.1'] = df['Cover.1'].str.replace('%','').astype(float)
df.head()
```

```python
# 绘制基础小提琴图
# Make boxplot for one group only
sns.violinplot(y=df["Cover"])
plt.show()
```

```python
sns.violinplot(y=df["Cover.1"])
plt.show()
```
