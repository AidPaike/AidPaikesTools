```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
DIR_COV = pd.read_csv("./data/DIE_cov.csv")
DIR_COV["name"] = "die"
all_frame = DIR_COV[["Cover.1","name"]][:-30000]
print(all_frame)
```

```python
Comfuzz_COV = pd.read_csv("./data/comfuzz_all.csv")
Comfuzz_COV["name"] = "comfort"
all_frame = pd.concat([all_frame, Comfuzz_COV[["Cover.1","name"]]], axis=0)
print(all_frame)
```

```python
fuzzilli_COV = pd.read_csv("./data/time_fuzzilli_interest_cov.csv")
fuzzilli_COV["name"] = "fuzzilli"
all_frame = pd.concat([all_frame, fuzzilli_COV[["Cover.1","name"]]], axis=0)
print(all_frame)
```

```python
codeAl_COV = pd.read_csv("./data/time_codeAl_data_15.csv")
codeAl_COV["name"] = "codeAl"
all_frame = pd.concat([all_frame, codeAl_COV[["Cover.1","name"]]], axis=0)
print(all_frame)
```

```python
"""
案例2：
绘制一个按分类变量分组的垂直小提琴图
"""
all_frame['Cover'] = all_frame['Cover.1'].str.replace('%','').astype(float)
sns.violinplot(x="name", y="Cover", data=all_frame ,width=1.5)
plt.ylim(22, 26)
plt.savefig('./images/result/result.png')#保存图片
plt.show()
```
